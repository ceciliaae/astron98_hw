```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1 How Many Plots?

def sines(x, A, w):
    plt.figure()
    plt.title("5 different sines")
    plt.xlabel("x data")
    plt.ylabel("y data")
    for i in range(np.size(A)):
        x = np.linspace(0, 2*np.pi, 1000)
        y = np.sin(x*w[i]) + A[i]
        plt.plot(x, y)
    plt.show
```

```python
#2 Randomness 

x = np.linspace(0, 41, 40)
list1 = np.random.randint(0, 100, size = (40))
list2 = np.random.randint(0, 100, size = (40))
plt.figure(figsize = (10,10))
plt.plot(x, list1, label = 'list1', color = "orange", linewidth = 10)
plt.plot(x, list2, label = 'list2', color = "red", linestyle="dashed")
plt.xlabel("index")
plt.ylabel("random number")
plt.legend(loc="lower left")
plt.show
```
